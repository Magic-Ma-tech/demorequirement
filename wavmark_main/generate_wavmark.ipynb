```python
import numpy as np
import soundfile
import torch
import wavmark
import os


test_dir = 'my_choose_test_50'
test_data = os.listdir(test_dir)
msg_path = './msg'
save_dir = 'wavmark_test_50'
model = wavmark.load_model().to('cuda')
for i in test_data:
    open_path = os.path.join(test_dir, i)

    payload = np.random.choice([0, 1], size=16)
    print("Payload:", payload)
    payload_tensor = torch.tensor(payload)
    torch.save(payload_tensor, f"{msg_path}/{i}_msg_tensor.pt")
    
    signal, sample_rate = soundfile.read(open_path)

    watermarked_signal, _ = wavmark.encode_watermark(model, signal, payload, show_progress=True)

    payload_decoded, _ = wavmark.decode_watermark(model, watermarked_signal, show_progress=True)
    BER = (payload != payload_decoded).mean() * 100

    soundfile.write(f'{save_dir}/wavmark_{i}', watermarked_signal, samplerate=sample_rate)

    print("Decode BER:%.1f" % BER)
```

```python
for i in test_data:
    watermarked_signal, sample_rate = soundfile.read(f'{save_dir}/wavmark_{i}')
    payload = torch.load(f"{msg_path}/{i}_msg_tensor.pt")
    payload_decoded, _ = wavmark.decode_watermark(model, watermarked_signal, show_progress=True)
    BER = (np.array(payload) != payload_decoded).mean() * 100

    print("Decode BER:%.1f" % BER)
```
